```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
x = diabetes_dataset.drop(columns = 'Outcome',axis=1)
y = diabetes_dataset['Outcome']
```

```python
scaler = StandardScaler()
scaler.fit(x)
standardized_data = scaler.transform(x)
```

```python
x = standardized_data
y = diabetes_dataset['Outcome']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, stratify= y, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(x_train,y_train)
```

```python
x_train_pred = classifier.predict(x_train)
acc_score = accuracy_score(x_train_pred,y_train)
print(acc_score*100)
```

```python
x_test_pred = classifier.predict(x_test)
acc_score = accuracy_score(x_test_pred,y_test)
print(acc_score)
```

```python
input_data = (4,110,92,0,0,37.5,0.191,30)
inp_arr = np.asarray(input_data)
input_data_reshaped = inp_arr.reshape(1,-1)

standardized_inp = scaler.transform(input_data_reshaped)
# print(standardized_inp)

prediction = classifier.predict(standardized_inp)
# print(prediction)

if(prediction[0]==0):
  print("Yayy!! You are not diabetic.")
else:
  print("Sorry You are diabetic.")
```
